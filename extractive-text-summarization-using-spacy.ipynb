```python
import spacy
import pytextrank
import urllib.request
import bs4 as bs
```

```python
nlp = spacy.load("en_core_web_lg")
```

```python
nlp.add_pipe("textrank")
```

```python
SOURCE = urllib.request.urlopen('https://en.wikipedia.org/wiki/Aircraft_in_fiction')


SOUP = bs.BeautifulSoup(SOURCE, 'lxml')

TEXT = ""
for paragraph in SOUP.find_all('p'):
    TEXT += paragraph.text
```

```python
TEXT
```

```python
doc = nlp(TEXT)
```

```python
for sent in doc._.textrank.summary(limit_sentences = 2):
    print(sent)
```

```python
for sent in doc._.textrank.summary(limit_phrases = 4, limit_sentences = 20):
    print(sent)
```

```python
phrases_and_ranks = [ 
    (phrase.chunks[0], phrase.rank) for phrase in doc._.phrases
]
phrases_and_ranks[:10]
```
